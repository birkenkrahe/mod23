### Difference equations



The constant growth model:

    results[t+1] = results[t] + annual_growth



#### Constant growth model



In mathematical notation, this is $x_{n+1} = x_{n} + c$, a difference
equation, which leads to $x_{n} = x_{0} + nc$. So if we know the initial
term and the constant growth factor, we know any value of the series.



#### Proportional growth model



This equation can be solved directly for a given value of n,
leading to a geometric progression $x<sub>0</sub>, x<sub>0</sub> &alpha;, x<sub>0</sub>&alpha;<sup>2</sup>, x<sub>0</sub>&alpha;<sup>3</sup>, x<sub>0</sub>&alpha;<sup>4</sup>,&#x2026;$
where $x_0$ is the first term of the series:

-   If $\alpha>1$< the sequence exhibits exponential growth.
-   If $0 < \alpha < 1$, it shows exponential decay.
-   If $\alpha<0$, the sequence will alternate in sign.

Plot the geometric series using Python for different values of $\alpha$:



```
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
x_0 = 1   # first term (previously a)
alphas = [1.5, 0.5, -0.5] # different values of alpha (previously r)
n = 10    # number of terms

# Create a subplot for each alpha value
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

for i, alpha in enumerate(alphas):
    # Generate geometric sequence
    geometric_sequence = x_0 * np.power(alpha, np.arange(n))
    # Plot
    axs[i].plot(geometric_sequence, marker='o', linestyle='-', color='blue')
    axs[i].set_xlabel('Term')
    axs[i].set_ylabel('Value')
    axs[i].set_title(f'Geometric Progression (alpha = {alpha})')
    axs[i].grid(True)

plt.tight_layout()
plt.savefig('img/geometricPython.png')
```

#### Quadratic model



This model, $x_{n+1} = x_{n} + r x_{n} ( 1 - \frac{x_{n}}{K})$ cannot be solved
analytically - it's nonlinear, recursive, and sensitive to initial
conditions - but we can approximate it with a differential equation
and solve that.

The code below plots the population over 100 iterations for different
values of $r$. You'll see that for lower values of $r$, the population
tends to stabilize, but as $r$ increases, particularly near 3.9, the
behavior becomes chaotic, showing no clear pattern or stability. This
illustrates the sensitivity to initial conditions and the
unpredictability inherent in chaotic systems.



```
import matplotlib.pyplot as plt
import numpy as np

def logistic_map(r, x, n_iter):
    """Returns the iterations of the logistic map."""
    iterations = [x]
    for _ in range(n_iter):
        x = x * r * (1 - x)
        iterations.append(x)
    return iterations

# Parameters
n_iter = 100
x_0 = 0.1  # Initial population

# Different values of r to illustrate various behaviors
r_values = [2.5, 3.2, 3.5, 3.9]

plt.clf()
plt.figure(figsize=(12, 8))

for r in r_values:
    result = logistic_map(r, x_0, n_iter)
    plt.plot(result, label=f'r = {r}')

plt.xlabel('Iteration')
plt.ylabel('Population')
plt.title('Logistic Map for Different Growth Rates (r)')
plt.legend()
plt.grid(True)
plt.savefig('img/chaos.png')
```
